## Gráficos de líneas.
Quizás el más sencillo de todos los gráficos que podamos realizar mediante el uso de **Matplotlib** corresponda a la visualización de una única función del tipo $y=f(x)$. Este será el primer ejercicio que realicemos oficialmente con esta librería. Para ello, y como en el resto de esta asignatura, inicializaremos nuestro script con el siguiente conjunto de instrucciones (donde intercambiaremos con frecuencia la plantilla de estilos que utilizaremos, a fin de ir acostumbrándonos de a poco a las alternativas que **Matplotlib** nos ofrece):

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Setting de plantilla tipo ggplot (ggplot2 es la librería de graficación de R).
plt.style.use("seaborn-white")
```

Para todos los gráficos que construyamos en **Matplotlib**, en general, siempre comenzaremos creando dos objetos: La **figura** propiamente tal (mediante la función `plt.figure()`), y los *ejes* que corresponderán a las referencias absolutas de nuestro gráfico (mediante la función `plt.axes()`). Siempre podremos controlar la resolución de nuestra figura al definirla, añadiendo el argumento `dpi` a la función `plt.figure()` y un valor numérico que define la correspondiente densidad de puntos por pixel:

```python
# La figura y los paneles que la limitan (en este caso, sólo uno).
fig = plt.figure(dpi=120)
ax = plt.axes()
```

En **Matplotlib**, la *figura* (la instancia que definimos mediante `plt.figure()`) puede ser pensada como un único contenedor que agrupa todos los objetos de Python que representan los ejes, gráficos, texto, anotaciones y etiquetasque están definidas para la imagen correspondiente. La instancia `ax` (resultante de la función `plt.axes()`) es lo que vemos en la imagen presentada en la lámina anterior: Una caja que limita la figura con etiquetas y marcadores (ticks), que eventualmente contendrá los elementos del gráfico que permitirán su visualización, y que puede ser más de uno en el caso de construir una grilla que contenga varios paneles de gráficos. En esta asignatura, en general, siempre referiremos estas instancias a las variables cuyos nombres son `fig` y `ax`, respectivamente.

Por cierto, cuando inicializamos una figura usando la función `plt.figure()`, podemos controlar sus dimensiones mediante el uso del parámetro `figsize`, el cual acepta una tupla con dos números de punto flotante que representan sus dimensiones en una relación de tipo `(largo, alto)`:

```python
fig = plt.figure(figsize=(8, 4), dpi=120)
ax = plt.axes();

plt.figure(figsize=(5, 5), dpi=120)
ax = plt.axes();

plt.figure(figsize=(3, 7), dpi=120)
ax = plt.axes();
```

Una vez creados los ejes, podemos utilizar la función `ax.plot()` para construir un sencillo **gráfico de líneas**. Comencemos pues con una función exponencial, a fin de ilustrar el procedimiento:

```python
fig = plt.figure(dpi=120) # Inicializamos nuestra figura.
ax = plt.axes() # Inicializamos el panel donde irá nuestro gráfico.
x = np.linspace(0, 3, 1000) # Construimos la data que irá en el eje horizontal.
ax.plot(x, np.exp(x)); # Construimos nuestro gráfico de líneas.
```

Alternativamente, podemos utilizar la interfaz tipo Matlab:

```python
plt.figure(dpi=120) # Inicializamos nuestra figura.
plt.plot(x, np.exp(x)); # Construimos nuestro gráfico de líneas.
```

Si queremos crear una única figura con múltiples gráficos de líneas, podemos simplemente llamar la función `plt.plot()` (o el método `plot()`, si usamos la interfaz orientada a objetos) tantas veces como queramos:

```python
plt.figure(dpi=120) # Inicializamos nuestra figura.
plt.plot(x, x**3 + 2*x**2 - 4*x + 1); # Construimos nuestro primer gráfico de líneas.
plt.plot(x, -2*x**3 - x**2 + x - 3); # Construimos nuestro segundo gráfico de líneas.
```

Y eso es todo lo que necesitamos saber para graficar rápidamente funciones en **Matplotlib**. Navegaremos en algunos detalles en breve, acerca de cómo controlar la apariencia de nuestro gráfico, en términos de los ejes y las líneas propiamente tales.

## Ajuste de colores y estilos.
El primer ajuste que podríamos querer realizar corresponde al control del color de las líneas y sus estilos. La función `plt.plot()` acepta argumentos adicionales para especificar ambas características de nuestros gráficos. Para ajustar el color, podemos usar el parámetro `color`, el cual acepta un parámetro de tipo string que representa, virtualmente, cualquier color que imaginemos. Y dicho color puede especificarse de una buena cantidad de formas:

```python
plt.figure(dpi=120) # Inicializamos nuestra figura.
x = np.linspace(start=0.1, stop=5, num=100) # Generamos nuestra data.

# Especificación del color por su nombre
plt.plot(x, np.log(x), color="green")

# Código corto de color, modelo sustractivo (rgbcmyk)
plt.plot(x, np.log(x) + 1, color="m")

# Escala de grises, entre 0 y 1.
plt.plot(x, np.log(x) + 2, color="0.7")

# Código hexadecimal (RRGGBB desde 00 a FF).
plt.plot(x, np.log(x) + 3, color="#FFDD44")

# Tupla RGB (modelo aditivo), valores entre 0 y 1.
plt.plot(x, np.log(x) + 4, color=(1.0, 0.2, 0.3))

# Código HTML de color.
plt.plot(x, np.log(x) + 5, color='chartreuse');
```

Si no se especifica ningún color, **Matplotlib** automáticamente ciclará entre una colección de colores por defecto para múltiples líneas en un gráfico, y que son dependientes del tipo de plantilla de estilos que utilicemos. Similarmente, el estilo de la línea puede setearse mediante el parámetro `linestyle`:

```python
# Definimos la data a graficar en el eje horizontal.
x = np.linspace(start=-2, stop=2, num=100)

# El tipo de línea se especifica mediante el prámetro linestyle.
plt.figure(dpi=120)
plt.plot(x, x**2 + 0, linestyle='solid')
plt.plot(x, x**2 + 1, linestyle='dashed')
plt.plot(x, x**2 + 2, linestyle='dashdot')
plt.plot(x, x**2 + 3, linestyle='dotted')

# El tipo de línea puede especificarse también mediante varios símbolos.
plt.plot(x, x**2 + 4, linestyle='-') # Equivalente a linestyle='solid'.
plt.plot(x, x**2 + 5, linestyle='--') # Equivalente a linestyle='dashed'.
plt.plot(x, x**2 + 6, linestyle='-.') # Equivalente a linestyle='dashdot'.
plt.plot(x, x**2 + 7, linestyle=':'); # Equivalente a linestyle='dotted'.
```

O bien, si deseamos compactar aún más nuestro código, podemos escribir:

```python
# Códigos compactos de línea y color.
plt.figure(dpi=120)
plt.plot(x, x**2 + 0, '-r') # Equivalente a linestyle="-", color="r".
plt.plot(x, x**2 + 1, '--c') # Equivalente a linestyle="--", color="c".
plt.plot(x, x**2 + 2, '-.m') # Equivalente a linestyle="-.", color="m".
plt.plot(x, x**2 + 3, ':g') # Equivalente a linestyle=":", color="g".
```

Los caracteres anteriores que permiten especificar los respectivos colores reflejan las abreviaturas de tipo estándar que suelen ser utilizadas en modelos de color típicos en el diseño de elementos gráficos, como es el caso del modelo aditivo (RGB) o sustractivo (CMYK) de colores.

## Límites de ejes de graficación.
En general, **Matplotlib** hace un trabajo bastante adecuado escogiendo los límites para los valores mostrados en cada eje de nuestros gráficos. Sin embargo, a veces es bueno tener el control de estos valores. La forma más directa de cambiarlos es mediante el uso de las funciones `plt.xlim()` y `plt.ylim()`, según sea el eje que deseemos intervenir:

```python
# Límites para los ejes de graficación.
plt.figure(dpi=120)
plt.plot(x, x**2, color="purple")
plt.xlim(-1, 1)
plt.ylim(-1, 1);
```

Si, por alguna razón, queremos que los valores mostrados en algún eje se muestren en orden inverso, bastará simplemente con especificar dicho orden en las funciones anteriores. Por ejemplo:

```python
# Inversión de la dirección del eje horizontal.
plt.figure(dpi=120)
plt.plot(x, x**2, color="purple")
plt.xlim(2, -2);
```

Una alternativa al setting anterior podría ser la utilización de la función `plt.axis()`, la que nos permite controlar todos los ejes de un gráfico al mismo tiempo, especificando como argumento una lista con una estructura de tipo `[xmin, xmax, ymin, ymax]`. Por ejemplo:

```python
# Uso de la función plt.axis() para controlar los límites de un eje.
plt.figure(dpi=120)
plt.plot(x, x**2, color="purple")
plt.axis([-1, 1, -1, 1]);
```

La función `plt.axis()` va mucho más allá. También nos permite hacer cosas tales como, automáticamente, ajustar al máximo los bordes de nuestra figura en torno al gráfico respectivo mediante el uso del parámetro `option`:

```python
# Uso de los tipos de ajuste de ejes.
plt.figure(dpi=120)
plt.plot(x, x**2, color="purple")
plt.axis("equal")
plt.title("equal");

plt.figure(dpi=120)
plt.plot(x, x**2, color="purple")
plt.axis("tight")
plt.title("tight");

plt.figure(dpi=120)
plt.plot(x, x**2, color="purple")
plt.axis("scaled")
plt.title("scaled");
```

## Rótulos de ejes, título y leyenda de un gráfico.
Para finalizar nuestros primeros pasos en la graficación con **Matplotlib**, exploraremos brevemente algunas opciones para rotular nuestras figuras. Esto es, agregar títulos a nuestros gráficos, rótulos de ejes y leyendas sencillas.

Los títulos y rótulos de ejes pueden especificar de manera muy sencilla mediante las funciones `plt.title()`, `plt.xlabel()` y `plt.ylabel()`. Tales rótulos son compatibles con código de tipo `TeX`, lo que nos permite incluso escribir ecuaciones o texto con subíndices o superíndices, siempre que permitamos que el intérprete de Python lea los símbolos `\`(*backslash*) de manera literal, lo que puede conseguirse anteponiendo una `r` antes del texto. Cualquier texto en `TeX` debe ir, dentro del string, entre dos símbolos `"$"`. Por ejemplo:

```python
# Un gráfico con sus rótulos de ejes y su título.
plt.figure(dpi=120)
plt.plot(x, x**2, color="purple")
plt.xlabel("x")
plt.ylabel(r"$x^{2}$")
plt.title(r"La función $y=x^{2}$");
```

La posición, tamaño y estilo de los rótulos de eje pueden ser ajustadas mediante varios argumentos. Por ejemplo, el parámetro `fontsize` nos permite especificar el **tamaño de fuente** de nuestro rótulo mediante un valor de tipo entero, mientras que el parámetro `fontweight` nos permite especificar el estilo de dichos rótulos mediante un string. Por ejemplo, `"bold"` para un rótulo en negrita, `"light"` para un estilo más suave. Finalmente, el parámetro `fontstyle` nos permite controlar el estilo de letra entre las opciones `normal` (por defecto) o `italic` (cursiva).

En caso de la dirección de estos rótulos, podemos incluso controlar su rotación respecto de su correspondiente eje de graficación mediante el argumento `rotation` (que es un entero entre 0 y 360, que define el ángulo de rotación), o su espaciamiento del mismo eje mediante `labelpad` (tipo entero):

```python
# Un gráfico con sus rótulos de ejes y su título. En este caso, los rótulos y el
# título vienen más personalizados.
plt.figure(dpi=120)
plt.plot(x, x**2, color="purple")
plt.xlabel("x", fontsize=12, labelpad=10, fontstyle="oblique")
plt.ylabel(r"$x^{2}$", fontsize=12, labelpad=12, rotation=0)
plt.title(r"La función $y=x^{2}$", fontsize=14, fontweight="bold");
```

Cuando se muestran múltiples líneas en un gráfico (o, usando la terminología de **Matplotlib**, en un mismo panel), puede ser útil crear una leyenda que especifique el significado de cada una de esas líneas. Y en este caso, **Matplotlib** también dispone de métodos para mostrar una leyenda en nuestras figuras, por ejemplo, mediante la función `plt.legend()`. Para ello, es necesario especificar además, en el constructor de cada gráfico, el rótulo de cada línea mediante el parámetro `label`, el cual corresponde a un argumento de la función `plt.plot()`:

```python
# Construcción de una leyenda sencilla.
plt.figure(dpi=120)
x = np.linspace(start=-5, stop=5, num=100)
plt.plot(x, np.sin(x), '-y', label=r'$sen(x)$')
plt.plot(x, np.cos(x), '-.g', label=r'$cos(x)$')
plt.legend();
```

Como resulta evidente, la función `plt.legend()` mantiene la trazabilidad tanto de los colores como del estilo de nuestras líneas. Cubriremos algunas opciones un poco más avanzadas a medida que vayamos aprendiendo a construir otros tipos de gráfico y personalizando sus distintos elementos.

## Equivalencias entre interfaces.
Mientras que la mayoría de las funciones de `pyplot` tienen métodos equivalentes y con el mismo nombre para el caso de la interfaz orientada a objetos, tales como `plt.plot()` y `ax.plot()`, o bien, `plt.legend()` y `ax.legend()`.

Sin embargo, otras funciones en la interfaz orientada a objetos sufren algunos cambios pequeños al transicionar entre ésta y la interfaz de tipo Matlab. Ejemplos de aquello son las correspondencias mostradas en la Tabla (1.1).

<p style="text-align: center;">Tabla (1.1): Equivalencias entre interfaces de Matplotlib</p>

| Función en interfaz de tipo Matlab | Método en interfaz orientada a objetos |
| :--------------------------------- | :------------------------------------- |
| `plt.xlabel()`                     | `ax.set_xlabel()`                      |
| `plt.ylabel()`                     | `ax.set_ylabel()`                      |
| `plt.xlim()`                       | `ax.set_xlim()`                        |
| `plt.ylim()`                       | `ax.set_ylim()`                        |
| `plt.title()`                      | `ax.set_title()`                       |

En la interfaz orientada a objetos, para la realización de gráficos, en vez de setear las componentes de nuestra figura una a una, a veces es más útil utilizar la función `ax.set()` para setear todas estas componentes al mismo tiempo:

```python
# Uso del setting mediante la función plt.axes().
fig = plt.figure(dpi=120)
ax = plt.axes()
ax.plot(x, np.sin(x))
ax.set(xlim=(0, 5), ylim=(-2, 2), xlabel='x', ylabel=r'$sen(x)$', title='Un gráfico sencillo');
```

## Gráficos de dispersión.
Otro tipo especial de gráfico que utilizamos muchísimo en labores comunes de ingeniería corresponde al **gráfico de dispersión**, que en la práctica es un *primo cercano* del gráfico de líneas. En vez de la utilización de puntos unidos mediante segmentos lineales o curvos, aquí dichos puntos son representados mediante círculos, triángulos, rectángulos u otras figuras personalizables a nuestro gusto.

### Uso de `plt.plot()`.
Anteriormente utilizamos las funciones `plt.plot()` o `ax.plot()` para generar gráficos de líneas. No obstante, dicha función igualmente puede utilizarse para generar gráficos de dispersión. Basta que especifiquemos el parámetro `marker`, el cual acepta un determinado string que representa un cierto tipo de marcador, pero dejando vacío el parámetro `linestyle`. Por ejemplo, para el caso de marcadores de tipo circulares, bastará con setear `marker="o"`:

```python
# Generamos algo de data.
x = np.linspace(start=0, stop=4*np.pi, num=50)
```

```python
# Construimos un sencillo gráfico de dispersión.
plt.figure(dpi=120)
plt.plot(x, np.cos(x), marker="o", color="black", linestyle="");
```

Existen muchos símbolos que permiten dibujar marcadores en un gráfico de dispersión. A continuación, mostraremos varios de ellos:

```python
# Definimos una semilla aleatoria fija.
rng = np.random.default_rng(42)
```

```python
# Mediante un loop, mostraremos varios de estos marcadores.
markers = ['o', '.', ',', 'x', '+', 'v', '^', '<', '>', 's', 'd']

plt.figure(figsize=(8, 5), dpi=120)
for marker in markers:
    plt.plot(rng.random(6), rng.random(6), marker, label="marker='{0}'".format(marker))
    plt.legend(loc="upper right")
    plt.xlim(0, 1.5);
```

Para ampliar nuestro abanico de posibilidades, estos códigos de caracteres pueden combinarse a fin de generar un plot que muestre un determinado estilo de líneas, puntos y colores para cada gráfico:

```python
# Combinamos estilos de líneas y marcadores.
plt.figure(dpi=120)
plt.plot(x, np.cos(x), "ob-");
```

Los argumentos adicionales para `plt.plot()` permiten especificar un enorme conjunto de opciones para líneas y marcadores:

```python
# Más opciones de personalización.
plt.figure(dpi=120)
plt.plot(
    x, np.cos(x), '-p', color='purple', markersize=15, linewidth=4, 
    markerfacecolor='white', markeredgecolor='purple', markeredgewidth=2
);
```

La flexibilidad de `plt.plot()`, como hemos podido observar, nos permite disponer de un enorme conjunto de posibles opciones para personalizar nuestros gráficos.
